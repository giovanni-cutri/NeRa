```python
# Updated data
models = ['RatingRGNN+Elo', 'Vanilla Elo', 'RatingRGNN+Berrar', 'RatingRGNN+Pi', 'RatingRGNN+no rating']
nbl_acc = [0.6102941176470589*100, 0.5488*100, 0.6176470588235294*100, 0.6213235294117647*100, 0.6176470588235294*100]
allbasket_acc = [58.49, 57.56, 58.49, 58.49, 52.97]
nfl_acc = [60.08, 60.73, 59.82, 59.82, 55.61]
soccer_acc = [48.10, 59.65, 48.10, 48.10, 48.58]

plt.rcParams.update({'font.size': 14})
# Helper function to create bar chart
def create_bar_chart(ax, accuracies, title, ylim):
    max_acc = max(accuracies)
    colors = ['b' if acc != max_acc else 'r' for acc in accuracies]
    ax.bar(models, accuracies, color=colors)
    ax.set_title(title, fontsize=16,)
    ax.set_ylabel('Validation Accuracy (%)', fontsize=14)
    ax.set_ylim(ylim)
    ax.tick_params(axis='x', labelsize=12)
    ax.tick_params(axis='y', labelsize=12)
    for tick in ax.get_xticklabels():
        tick.set_rotation(45)


        
# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# NBL Bar Chart
create_bar_chart(axs[0, 0], nbl_acc, 'NBL Validation Accuracies', [50, 70])

# All Basketball Bar Chart
create_bar_chart(axs[0, 1], allbasket_acc, 'All Basketball Validation Accuracies', [50, 70])

# NFL Bar Chart
create_bar_chart(axs[1, 0], nfl_acc, 'NFL Validation Accuracies', [50, 70])

# Soccer Bar Chart
create_bar_chart(axs[1, 1], soccer_acc, 'Soccer Validation Accuracies', [40, 70])

# Adjust layout
plt.tight_layout()
plt.show()
```
